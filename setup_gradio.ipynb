```python
import os
from PIL import Image

def run(cloth_img, person_img):
    save_inputs(cloth_img, person_img)

    # Run the model (replace with actual working command from the repo)
    os.system("python test.py --name vitonhd_test --datasetting test --data_list test_pairs.txt --stage TOM --checkpoint checkpoints/TOM_final.pth")

    output_file = "/content/results/test/TOM/00001_00.jpg"
    if os.path.exists(output_file):
        return Image.open(output_file)
    else:
        raise FileNotFoundError("Try-on model failed: Output image not found.")
```

```python
import os

def save_inputs(cloth_img, person_img):
    base_path = "/content/inputs/test/"

    # Make required folders
    folders = ["cloth", "image", "cloth-mask", "image-parse", "openpose-img", "openpose-json"]
    for folder in folders:
        os.makedirs(os.path.join(base_path, folder), exist_ok=True)

    # Save the uploaded images
    cloth_path = os.path.join(base_path, "cloth", "00001_00.jpg")
    person_path = os.path.join(base_path, "image", "00001_00.jpg")

    cloth_img.save(cloth_path)
    person_img.save(person_path)

    # ðŸ’¡ Optional: Create dummy files for required additional inputs
    # You can replace these with real preprocessing later
    dummy_image = cloth_img.resize((768, 1024))  # Match model expected input size
    dummy_image.save(os.path.join(base_path, "openpose-img", "00001_00_rendered.png"))
    dummy_image.save(os.path.join(base_path, "cloth-mask", "00001_00.png"))
    dummy_image.save(os.path.join(base_path, "image-parse", "00001_00.png"))

    # Create an empty/dummy OpenPose JSON file
    dummy_json = '{"people": [{"pose_keypoints_2d": [0]*54}]}'
    with open(os.path.join(base_path, "openpose-json", "00001_00_keypoints.json"), "w") as f:
        f.write(dummy_json)

    print("âœ… Images and dummy files saved.")
```

```python
import gradio as gr

with gr.Blocks() as app:
    with gr.Row():
        gr.Markdown("<center><h1> Clothes Virtual Try ON </h1><center>")
    with gr.Row():
        with gr.Column():
            cloth_input = gr.Image(sources=['upload'], type="pil", label="Upload the Cloth Image")
        with gr.Column():
            model_input = gr.Image(sources=['upload'], type="pil", label="Upload the Human Image")
    with gr.Row():
        final_output = gr.Image(sources=['upload'], type="pil", label="Final Prediction")
    with gr.Row():
        submit_button = gr.Button("Submit")

    submit_button.click(fn=run, inputs=[cloth_input, model_input],
                        outputs=[final_output])
app.launch(debug=True, share=True)
```
